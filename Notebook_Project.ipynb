```python
import pandas as pd
import numpy as np
from faker import Faker
import random
import datetime

# Cria um Faker para gerar dados fictícios
fake = Faker()

# Define o número de registros a serem gerados
num_records = 100000

# Gera dados para o dataframe
data = {
    'ID_Transacao': [fake.uuid4() for _ in range(num_records)],
    'ID_Produto': [random.randint(1, 100) for _ in range(num_records)],
    'ID_Cliente': [fake.uuid4() if random.random() > 0.2 else np.nan for _ in range(num_records)],
    'Quantidade (Unid.)': [random.randint(1, 5) for _ in range(num_records)],
    'Preco_Unitario ($)': [round(random.uniform(10, 1000), 2) if random.random() > 0.1 else np.nan for _ in range(num_records)],
    'Data_Transacao': [fake.date_time_between(start_date='-1y', end_date='now') for _ in range(num_records)],
    'Latitude_Origem': [random.uniform(-90, 90) for _ in range(num_records)],  # Coordenadas geográficas da origem
    'Longitude_Origem': [random.uniform(-180, 180) for _ in range(num_records)],
    'Latitude_Destino': [random.uniform(-90, 90) for _ in range(num_records)],  # Coordenadas geográficas do destino
    'Longitude_Destino': [random.uniform(-180, 180) for _ in range(num_records)],
    'Tempo_Entrega_Estimado (h)': [random.randint(1, 24) for _ in range(num_records)],  # Tempo estimado de entrega em horas
    'Prioridade_Entrega': [random.randint(1, 5) for _ in range(num_records)]  # Prioridade da entrega
}

# Cria o dataframe
df = pd.DataFrame(data)
```

```python
import great_expectations as gx
context = gx.get_context()
validator = context.sources.pandas_default.read_csv(
    "/content/Rotas_de_entrega_em_ecommerce.csv"
)
```

```python
df.columns
```

```python
# Validando a ausência de valores nulos em colunas específicas
columns_to_check_for_null = [
    'ID_Transacao', 'ID_Produto', 'ID_Cliente',
    'Quantidade (Unid.)', 'Preco_Unitario ($)',
    'Data_Transacao', 'Latitude_Origem', 'Longitude_Origem',
    'Latitude_Destino', 'Longitude_Destino',
    'Tempo_Entrega_Estimado (h)', 'Prioridade_Entrega'
]

for column in columns_to_check_for_null:
    validator.expect_column_values_to_not_be_null(column)

validator.expect_column_values_to_be_between(
    'Quantidade (Unid.)', min_value=1
)
validator.expect_column_values_to_be_between(
    'Tempo_Entrega_Estimado (h)', min_value=1
)

validator.expect_column_values_to_be_between(
    'Preco_Unitario ($)', min_value=1
)

validator.save_expectation_suite(discard_failed_expectations=False)
```

```python
checkpoint = context.add_or_update_checkpoint(
    name="my_quickstart_checkpoint",
    validator=validator,
)
```

```python
checkpoint_result = checkpoint.run()
```

```python
context.view_validation_result(checkpoint_result)
```
